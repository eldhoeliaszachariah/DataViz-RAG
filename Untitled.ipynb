The RAG Workflow & Pipeline
Before writing any code, it is crucial to understand the step-by-step pipeline of how a Retrieval-Augmented Generation (RAG) application works. When you click "Process" in the app, the following 5-step pipeline executes:

1. Ingestion (Document Loading)

Input: User uploads PDF files.
Action: The app opens the PDFs and extracts all the raw text, combining it into one giant string.
2. Chunking (Text Splitting)

Input: The giant string of raw text.
Action: Because LLMs have context limits and search databases work better with small snippets, the app chops the text into 1000-character "chunks". We overlap them by 200 characters so we don't accidentally cut a sentence in half.
3. Embedding (Vectorization)

Input: The list of text chunks.
Action: This is the "magic" step. We pass each chunk through an Embedding Model. This model reads the text and converts it into a long array of numbers (a vector). Texts with similar meanings get numbers that are mathematically close to each other.
4. Indexing (Storing in FAISS)

Input: The list of numerical vectors.
Action: We store these vectors in a FAISS database. FAISS organizes these numbers in a way that allows lightning-fast mathematical searching.
5. Retrieval & Generation (The Chat Loop)

Input: User asks a question (e.g., "What is the summary of chapter 1?").
Action:

The app converts the user's question into a vector using the same Embedding Model.
FAISS searches its database and returns the top 3-4 chunks that have numbers closest to the question.
The app sends the User Question + the Retrieved Document Chunks to the LLM (Llama 3.2).

Llama 3.2 reads the chunks and generates a final, human-readable answer.

Explanation of Each Tool
Below is a breakdown of every major library used in this project and how it fits into the pipeline above.

1. Streamlit
Streamlit is a Python framework used to quickly build interactive web apps for ML/AI without frontend coding.

Use in this project: It acts as the entire User Interface (UI)—rendering the chat window, handling PDF uploads, and displaying responses.
2. Markdown (Python markdown library)
Markdown is a simple formatting language for text (headers, lists, bold).

Use in this project: LLMs output Markdown. We use this library to convert that Markdown into HTML so it renders beautifully inside our custom chat bubbles.

3. PyPDF2
PyPDF2 is a pure-Python library used to read and extract text from PDF files.

Use in this project: It handles Step 1 (Ingestion). It pulls raw text out of the uploaded PDFs.

4. Ollama (ollama Python library)
Ollama lets you run Large Language Models (LLMs) locally on your machine.

Use in this project: It handles Step 5 (Generation). It runs llama3.2 locally to read retrieved document chunks and generate answers, without needing cloud APIs.

5. LangChain (langchain)
LangChain is a framework providing standard interfaces for chains, memory, prompts, and retrievers.

Use in this project: It is the "glue". It ties the Ollama LLM, the FAISS database, the custom prompts, and the conversation memory together into a single automated pipeline (ConversationalRetrievalChain).

6. HuggingFace Embeddings (sentence-transformers)
This library provides models that convert human text into arrays of numbers (vectors).

Use in this project: It handles Step 3 (Embedding). We use all-MiniLM-L6-v2 to translate text chunks into numerical vectors.

7. FAISS (faiss-cpu)
FAISS (Facebook AI Similarity Search) is a library for fast similarity searching of dense vectors.

Use in this project: It handles Step 4 (Indexing) and part of Step 5 (Retrieval). It stores the vectors and instantly finds chunks matching the user's question.

8. PyTorch (torch)
PyTorch is the underlying deep learning framework.

Use in this project: We use torch.cuda.is_available() to automatically detect if the user has an NVIDIA GPU. If yes, the HuggingFace embedding model runs on the GPU (much faster). If no, it falls back to the CPU.

9. ConversationBufferMemory (LangChain)
A module that temporarily stores the back-and-forth history of a chat session.

Use in this project: Without memory, the bot wouldn't know what "it" means in a follow-up question. This saves previous context for the LLM.

CONFIGURATION & SETUP
Before writing the logic, we configure the environment:

Hardware Auto-Detection: Checking for a GPU (CUDA).
Model Naming: Declaring the LLM and Embedding model names.
Custom Prompting: Creating a template to rewrite follow-up questions (e.g., turning "summarize it" into "summarize the document about black holes") before searching the database.

```python
#========= IMPORTS FOR CONFIGURATION ===========
import torch                                           # PyTorch library used to check hardware capabilities
from langchain.prompts import PromptTemplate           # Used to format our custom instruction string

#========= HARDWARE AUTO-DETECTION ===========
# Check if an NVIDIA GPU (CUDA) is available on the machine
# If True, use "cuda". If False (no GPU found), fallback to "cpu"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Print the result so the developer knows what hardware is being utilized
print(f"🚀 System detected. Using device: {DEVICE} for embeddings.")

#========= MODEL CONFIGURATION ===========
# Define the exact name of the local model Ollama will use to generate text
LLM_MODEL = "llama3.2"                                 

# Define the exact name of the HuggingFace model used to create vector embeddings
EMBEDDING_MODEL = "sentence-transformers/all-MiniLM-L6-v2" 

#========= CUSTOM PROMPT CONFIGURATION ===========
# Write a specific instruction telling the AI how to handle follow-up questions
custom_template = """Given the following conversation and a follow up question, rephrase the follow up question to be a standalone question, in its original language.
Chat History:
{chat_history}
Follow Up Input: {question}
Standalone question:"""

# Convert the plain text string into a Langchain PromptTemplate object
# This allows Langchain to dynamically inject the 'chat_history' and 'question' later
CUSTOM_QUESTION_PROMPT = PromptTemplate.from_template(custom_template)
```

DATA PROCESSING FUNCTIONS (Ingestion & Chunking)
These functions handle Steps 1 and 2 of the pipeline:

get_pdf_text: Opens PDFs and pulls out text into one giant string.
get_chunks: Chops the giant string into 1000-character pieces, overlapping by 200 characters to preserve context.

```python
#========= IMPORTS FOR DATA PROCESSING ===========
from PyPDF2 import PdfReader                           # Library to read PDF files
from langchain.text_splitter import CharacterTextSplitter # Library to chop text into pieces

#========= FUNCTION: EXTRACT TEXT ===========
def get_pdf_text(docs):
    """Takes a list of uploaded PDF files and returns a single string containing all text."""
    text = ""                                          # Initialize an empty string to hold everything
    for pdf in docs:                                   # Loop through every PDF the user uploaded
        pdf_reader = PdfReader(pdf)                    # Open the PDF using PyPDF2
        for page in pdf_reader.pages:                  # Loop through every single page in the current PDF
            text += page.extract_text()                # Extract the text and append it to our string
    return text                                        # Return the massive combined text string

#========= FUNCTION: SPLIT TEXT ===========
def get_chunks(raw_text):
    """Takes a large string of text and splits it into smaller, overlapping chunks."""
    # Configure the splitter rules
    text_splitter = CharacterTextSplitter(
        separator="\n",                                # Prefer to split text at line breaks
        chunk_size=1000,                               # Maximum length of each chunk (1000 characters)
        chunk_overlap=200,                             # Overlap chunks by 200 chars to preserve context
        length_function=len                            # Use standard Python len() to count characters
    )   
    # Execute the splitting on the raw text
    chunks = text_splitter.split_text(raw_text)        
    return chunks                                      # Return a list of text chunks
```

VECTOR STORE CREATION (Embedding & Indexing)
This handles Steps 3 and 4 of the pipeline. We take text chunks and pass them through the HuggingFace embedding model. This translates the human-readable text into arrays of numbers (vectors). We then load these vectors into a FAISS database.

```python
#========= IMPORTS FOR VECTORIZING ===========
from langchain.embeddings import HuggingFaceEmbeddings # Creates the numerical vectors
from langchain.vectorstores import faiss               # The FAISS database to store the vectors

#========= FUNCTION: CREATE VECTOR STORE ===========
def get_vectorstore(chunks):
    """Converts text chunks into vector embeddings and stores them in FAISS."""
    # Initialize the embedding model, passing the auto-detected DEVICE (GPU or CPU)
    embeddings = HuggingFaceEmbeddings(
        model_name=EMBEDDING_MODEL,                    # Use the model defined in config
        model_kwargs={'device': DEVICE}                # Force it to use GPU if available, else CPU
    )
    
    # Create a FAISS vector store directly from the text chunks and embeddings
    # FAISS handles the indexing automatically in the background
    vectorstore = faiss.FAISS.from_texts(texts=chunks, embedding=embeddings)
    
    return vectorstore, embeddings                     # Return both the database AND the embedding model object
```

LLM & CHAIN SETUP (Generation & Memory)
Here we build Step 5 of the pipeline using LangChain's ConversationalRetrievalChain. This chain acts as an automated assembly line: It takes a question -> uses Memory to rewrite it -> searches FAISS -> passes the retrieved chunks to Llama 3.2 -> generates the final answer.

```python
#========= IMPORTS FOR CHAIN SETUP ===========
from langchain.llms import Ollama                      # The local LLM interface
from langchain.memory import ConversationBufferMemory  # The chat history storage
from langchain.chains import ConversationalRetrievalChain # The main orchestration chain

#========= FUNCTION: BUILD CONVERSATION CHAIN ===========
def get_conversationchain(vectorstore):
    """Connects the LLM, the Vector Store, Memory, and Prompts into one unified chain."""
    # Initialize the local Llama 3.2 model via Ollama
    llm = Ollama(model=LLM_MODEL)                     
    
    # Set up the memory module to remember the conversation
    memory = ConversationBufferMemory(
        memory_key='chat_history',                     # The variable name Langchain uses for history
        return_messages=True,                          # Return full message objects (needed for our UI)
        output_key='answer'                            # Tell memory where the final answer lives
    )
    
    # Build and return the main chain
    conversation_chain = ConversationalRetrievalChain.from_llm(
        llm=llm,                                       # The generator (Llama 3.2)
        retriever=vectorstore.as_retriever(),          # The searcher (FAISS)
        condense_question_prompt=CUSTOM_QUESTION_PROMPT,# The re-phraser (our custom prompt)
        memory=memory                                  # The context holder (ConversationBufferMemory)
    )
    return conversation_chain                          
```

MAIN APP EXECUTION (Streamlit UI)
Note: The code below belongs in a file named app.py. Streamlit scripts are meant to be run from the terminal via streamlit run app.py, not directly inside a Jupyter Notebook cell, because they create a continuous web server loop.

This cell ties all the previously defined functions to the Streamlit UI. It handles displaying chat messages (including fixing the duplication bug and rendering Markdown properly).

```python
#========= IMPORTS FOR MAIN APP ===========
import markdown                                        # To convert LLM markdown to HTML
from dotenv import load_dotenv                         # Load environment variables
import streamlit as st                                 # The web framework
from htmlTemplates import css, bot_template, user_template # Our custom UI styling

#========= MAIN FUNCTION ===========
def main():
    # Load .env variables (good practice, even if mostly local now)
    load_dotenv()
    
    # Configure the Streamlit webpage settings
    st.set_page_config(page_title="Chat with multiple PDFs (Local Llama3.2)", page_icon=":books:", layout="wide")
    
    # Inject our custom CSS into the Streamlit app
    st.write(css, unsafe_allow_html=True)
    
    # Initialize Streamlit session state variables to persist data across page reloads
    if "conversation" not in st.session_state:
        st.session_state.conversation = None           # Will hold our Langchain chain
    if "chat_history" not in st.session_state:
        st.session_state.chat_history = None           # Will hold the list of messages
    
    # Render the main page header and text input box
    st.header("Chat with multiple PDFs :books:")
    question = st.text_input("Ask a question about your documents:")
    
    # If the user typed a question AND the chain has been built, process it
    if question and st.session_state.conversation is not None:
        # 1. Invoke the chain with the user's question
        response = st.session_state.conversation({'question': question})
        
        # 2. Save the updated history back to session state
        st.session_state.chat_history = response["chat_history"]
        
        # 3. Extract ONLY the last two messages (User & Bot) to prevent duplicating the whole chat on screen
        user_msg = st.session_state.chat_history[-2]
        bot_msg = st.session_state.chat_history[-1]
        
        # 4. Format User message (convert newlines to HTML breaks) and render
        safe_user_text = user_msg.content.replace('\n', '<br>')
        st.write(user_template.replace("{{MSG}}", safe_user_text), unsafe_allow_html=True)
        
        # 5. Format Bot message (convert Markdown to HTML) and render
        html_bot_text = markdown.markdown(bot_msg.content)
        st.write(bot_template.replace("{{MSG}}", html_bot_text), unsafe_allow_html=True)
        
    # ========== SIDEBAR FOR DOCUMENT UPLOAD ==========
    with st.sidebar:
        st.subheader("Your documents")
        # Create a file uploader widget accepting multiple PDFs
        docs = st.file_uploader("Upload your PDFs here and click 'Process'", accept_multiple_files=True)
        
        # When the user clicks the Process button
        if st.button("Process"):
            # Ensure they actually uploaded something
            if docs is None:
                st.warning("Please upload at least one PDF file.")
            else:
                with st.spinner("Processing"):
                    # Step 1: Extract text from PDFs
                    raw_text = get_pdf_text(docs)
                    
                    # Step 2: Split text into chunks
                    text_chunks = get_chunks(raw_text)
                    
                    # Step 3: Create vector store (embeddings + FAISS)
                    vectorstore = get_vectorstore(text_chunks)
                    
                    # Step 4: Create the conversation chain and save it to session state
                    st.session_state.conversation = get_conversationchain(vectorstore)
                    
                    st.success("Documents processed successfully! You can now ask questions.")

# Standard Python pattern to run the script
if __name__ == '__main__':
    main()

# ---------------------------------------------------------
# INSTRUCTIONS: 
# Copy everything from Cell 5 to this current cell, 
# save it in a file called `app.py` in the same folder as 
# your `htmlTemplates.py`, and run in your terminal:
# streamlit run app.py
# ---------------------------------------------------------
```
